```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.style as style
style.use('ggplot')
```

```python
df = pd.read_csv('robinhood_data/100_most_popular.csv')
df
```

```python
df['begins_at'] = df['begins_at'].str.split('T', expand=True)
```

```python
df['begins_at']
```

```python
df['Date'] = pd.to_datetime(df['begins_at'])
```

```python
df.set_index(df['Date'], inplace=True)
df
```

```python
df.drop(['begins_at', 'Date'], axis=1, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
df.to_csv('robinhood_data/100_most_popular_cleaned.csv')
```

```python
# Plotting Alibaba Stocks Closed Prices past five years 
plt.figure(figsize=(22,10))
plt.title('Closed Prices')
plt.plot(df.index, df['BABA'], color='green', label='Alibaba')
plt.legend()
plt.show()
```
