Scrapping the S&P 500 list from wikipedia

```python
# scrapping data from this site
# https://en.wikipedia.org/wiki/List_of_S%26P_500_companies

import bs4 as bs # for formatting and webscrapping 
import pickle # serialize any python object, saving variables etc
import requests # processes get request from site 

import datetime as dt
import os
```

```python
def save_sp500_tickers():
    resp = requests.get('https://en.wikipedia.org/wiki/List_of_S%26P_500_companies') # get requests to website
    soup = bs.BeautifulSoup(resp.text, 'lxml') # convert to BeautifulSoup object
    table = soup.find('table', {'class': 'wikitable sortable'}) # locate table by class name wikitable sortable
    print(table)
    tickers = []

        # iterate through the table rows, disregard the header
    for row in table.findAll('tr')[1:]:
#         print(row)
        ticker = row.findAll('td')[0].text # get the first column and convert to str object this is where the names are
#         print(ticker)
        ticker_ = ticker.split('\n')
#         print(ticker[0])
        tickers.append(ticker_[0])
#     print(tickers)
    
    with open('sp500tickers.pkl', 'wb') as f: # writing bytes to new file
        pickle.dump(tickers, f) # dumping tickers to file f
    
#     print(tickers)
    return tickers
    
```

```python
save_sp500_tickers()
```

```python
def get_data_from_yahoo(reload_sp500=False):
    if reload_sp500:
        tickers = save_sp500_tickers()
    
    else:
         with open('sp500tickers.pkl', 'rb') as f: # writing bytes to new file
            tickers = pickle.load(f)
            
    if not os.path.exists('stock_dfs'):
        os.makedirs('stock_dfs')
    
    start = dt.datetime(2000, 1, 1)
    end = dt.datetime.now()
    
    for ticker in tickers:
        print(ticker)
        
        if not os.path.exists('stock_dfs/{}.csv'.format(ticker)):
            df = web.DataReader(ticker, 'yahoo', start, end)
            df.to_csv('stock_dfs/{}.csv'.format(ticker))
            
        else:
            print('Already have: {}'.format(ticker))
```

```python
get_data_from_yahoo()
```

Next We will begin  Taking Adjusted Close Price for all csv files and combing that into a single dataframe, each column is a seperate company

```python
df = pd.read_csv('stock_dfs/{}.csv'.format('A'))
df.columns
```

```python
def compile_data():
    
    with open('sp500tickers.pkl', 'rb') as f:
        tickers = pickle.load(f)
        
    main_df = pd.DataFrame()
    
    for count, ticker in enumerate(tickers[:66]):
#     for count, ticker in enumerate(tickers):
        df = pd.read_csv('stock_dfs/{}.csv'.format(ticker))
        df.set_index('Date', inplace=True)
        df.rename(columns={'Adj Close': ticker}, inplace=True)
        df.drop(['Open','High' ,'Low' ,'Close', 'Volume'], axis=1, inplace=True)
        
        if main_df.empty:
            main_df = df
            
        else:
            main_df = main_df.join(df, how='outer')
            
        if count % 10 == 0:
            print(count)
    
    main_df.to_csv('sp500_joined_closes.csv')
        
```

```python
compile_data()
```
