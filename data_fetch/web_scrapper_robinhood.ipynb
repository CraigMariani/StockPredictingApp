```python
## getting the tickers of the 100 most popular list off of Robinhood 

resp = requests.get('https://robinhood.com/collections/100-most-popular')
# soup = bs.BeautifulSoup(resp.text, 'lxml') # parse into soup object
soup = bs.BeautifulSoup(resp.content, 'html.parser')
table = soup.find("table", {"class": "_25eGUhpRN5n6jaWqzmSOZl table"})
```

```python
tickers = []

for row in table.findAll('tr')[1:]:
    ticker = row.findAll('td')[1].text
    tickers.append(ticker)
```

```python
tickers.sort()
tickers
```

```python
names = []

for row in table.findAll('tr')[1:]:
    name = row.findAll('td')[0].text
    names.append(name)
```

```python
# converting ticker list to csv
df_tickers = pd.DataFrame(tickers)
df_tickers.columns = ['TickerLabel']
df_tickers.to_csv('robinhood_data/ticker_lists/100_most_popular.csv')
```

```python
## USING ROBINHOOD API TO GET 5 years of historical data

for index, ticker in enumerate(tickers):
    ticker_str = str(ticker)
    data = r.stocks.get_stock_historicals(ticker_str,interval='day', span='5year', bounds='regular')
    df = pd.DataFrame(data)
    df.to_csv('robinhood_data/stock_spreadsheets/{}_{}.csv'.format(ticker_str, names[index]))
```

```python
main_df = pd.DataFrame()

for count, ticker in enumerate(df_tickers['TickerLabel']):
    df = pd.read_csv('robinhood_data/stock_spreadsheets/{}_{}.csv'.format(ticker, 
                                                                          df_tickers['Name'][count]))
    df.drop(['Unnamed: 0'], axis=1, inplace=True)
    df.set_index('begins_at', inplace=True)
    df.rename(columns={'close_price': ticker}, inplace=True)
    df.drop(['open_price', 'high_price', 'low_price', 'volume',
             'session', 'interpolated', 'symbol'], 
            axis=1, inplace=True)
    
    if main_df.empty:
        main_df = df
    
    else:
        main_df = main_df.join(df, how='outer')
        
    main_df.to_csv('robinhood_data/100_most_popular.csv')
    
```
